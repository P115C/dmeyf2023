```python
%%sql

create or replace table competencia_01 as 

select
    *,
    lead(t0.foto_mes_ts, 1) over ventana as foto_mes_ts_prox_1, 
    lead(t0.foto_mes_ts, 2) over ventana as foto_mes_ts_prox_2
from competencia_01 t0
window ventana as (partition by numero_de_cliente order by foto_mes)
```

```python
%%sql

select 
    *
from competencia_01
limit 5
```

```python
%%sql

create or replace table competencia_01 as 

select
    *,
    case
        WHEN (date_diff('month', t0.foto_mes_ts, TIMESTAMP '2021-05-01') <= 1) THEN NULL
        WHEN ((t0.foto_mes_ts_prox_1 is NULL) OR 
              (date_diff('month', t0.foto_mes_ts, t0.foto_mes_ts_prox_1) != 1)) THEN 'BAJA+1'
        WHEN ((t0.foto_mes_ts_prox_2 is NULL) OR 
              (date_diff('month', t0.foto_mes_ts, t0.foto_mes_ts_prox_2) != 2)) THEN 'BAJA+2' 
        ELSE 'CONTINUA'
    END as clase_ternaria
from competencia_01 t0
```

```python
%%sql

ALTER TABLE competencia_01 DROP foto_mes_ts;
ALTER TABLE competencia_01 DROP foto_mes_ts_prox_1;
ALTER TABLE competencia_01 DROP foto_mes_ts_prox_2;
```

```python
%%sql

select 
    clase_ternaria, 
    count(*) as cant,
from competencia_01
group by clase_ternaria
```

## Ejemplo (funciones analíticas)

```python
%%sql

create or replace table ejemplo(foto_mes int, nombre VARCHAR, viyuya float);

insert into ejemplo values
(202101, 'Gustavo', 1000)
, (202102, 'Gustavo', 1100)
, (202103, 'Gustavo', 1200)
, (202101, 'Joaquin', 10000)
, (202102, 'Joaquin', 20000)
, (202103, 'Joaquin', 30000)
, (202101, 'Ale', -100)
, (202102, 'Ale', -null)
, (202103, 'Ale', -300);

SELECT *
FROM ejemplo
LIMIT 5
```

```python
%%sql

select
    *
    , lead(viyuya, 1) over (partition by nombre order by foto_mes) as viyuya_mas_1
    , lag(viyuya, 1) over (partition by nombre order by foto_mes) as viyuya_menos_1
    , lag(viyuya, 2) over (partition by nombre order by foto_mes) as viyuya_menos_2
from ejemplo
```
